```python
def anagram_checker(word_a, word_b):
    # Normalize: convert to lowercase
    word_a = word_a.lower()
    word_b = word_b.lower()
    
    # ORemove spaces 
    word_a = word_a.replace(" ", "")
    word_b = word_b.replace(" ", "")
    
    # Check if sorted characters match
    return sorted(word_a) == sorted(word_b)
```

```python
# mroe efficient version
from collections import Counter

def anagram_checker(str1, str2):
    return Counter(str1.lower()) == Counter(str2.lower())
```

```python
# Example tests
print(anagram_checker("Silent", "listen"))  # True
print(anagram_checker("Silent", "Night"))   # False
print(anagram_checker("night", "Thing"))    # True
```

# Assignment 1 - Part II

```python
def anagram_checker(word_a, word_b, is_case_sensitive=False):
    # Handle case sensitivity
    if not is_case_sensitive:
        word_a = word_a.lower()
        word_b = word_b.lower()
    
    # Optional: remove spaces (in case phrases are used later)
    word_a = word_a.replace(" ", "")
    word_b = word_b.replace(" ", "")
    
    # Check if sorted characters match
    return sorted(word_a) == sorted(word_b)
```

```python

# Example tests
print(anagram_checker("Silent", "listen", False)) # True

print(anagram_checker("Silent", "listen", True))     # False (case-sensitive)
print(anagram_checker("night", "Thing", True))       # False
```

1. Create a class called `Book` that has the attributes title, author, pages, price. It also has the method get_price() which returns price, and set_price() to ensure price is numeric.

Then create two instances of this class:

b1:
* title -> 'Practical Programming'
* author -> 'Gries, Campbell, Montojo'
* pages -> 383
* price -> 50

b2:
* title -> 'Building a Career in Data Science'
* author -> 'Robinson, Nolis'
* pages -> 322
* price -> 40

```python
class Book:
    def __init__(self, title, author, pages, price):
        self.title = title
        self.author = author
        self.pages = pages
        self.__price = price
    def get_price(self):
        return self.__price
    def set_price(self, new_price):
        if type(new_price) == int or type(new_price) == float:
            self.__price = new_price
        else:
            print("Invalid price format")
    def total_price(self):
        return(round(self.__price * 1.13),2)
```
